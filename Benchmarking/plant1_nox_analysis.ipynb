# Post optimizatioin

```python
# Replace missing values with avg vals
db0 = pd.read_csv('/media/m3rg2000/mounted/Junaid/.../Emissions/Old emission/db0_kiln.csv')
df = db0[['NOx content in the raw gas (preheater outlet)', 'Clinker production']]
df.fillna(df.mean(numeric_only=True), inplace=True)
df['NOx content in the raw gas (preheater outlet)'] = np.where(
    df['NOx content in the raw gas (preheater outlet)'] > 500,
    df['NOx content in the raw gas (preheater outlet)'] * (1-0.64),
    df['NOx content in the raw gas (preheater outlet)'] * (1-0.34)
)

df['NOx content in the raw gas (preheater outlet)'] = df['NOx content in the raw gas (preheater outlet)']*2.05 # NOx ppm to mg/Nm3
df['Clinker production'] = df['Clinker production']*16.667 # tph to kg clk/min
df['mg NOx/min'] = df['NOx content in the raw gas (preheater outlet)'] *df['Clinker production']  *0.55 #(0.55m3/kg clk). Net unit mg NOx/min
```
